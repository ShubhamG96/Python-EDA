```python
Matches['city']=Matches['city'].fillna(0)
Matches['winner']=Matches['winner'].fillna(0)
Matches['player_of_match']=Matches['player_of_match'].fillna(0)
Matches['umpire1']=Matches['umpire1'].fillna(0)
Matches['umpire2']=Matches['umpire2'].fillna(0)
Matches['umpire3']=Matches['umpire3'].fillna(0)
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(Matches.isnull())
```

```python
Matches.info()
```

```python
#Converting data Type
Matches['city']=Matches['city'].astype(str)
Matches['team1']=Matches['team1'].astype(str)
Matches['team2']=Matches['team2'].astype(str)
Matches['toss_winner']=Matches['toss_winner'].astype(str)
Matches['toss_decision']=Matches['toss_decision'].astype(str)
Matches['result']=Matches['result'].astype(str)
Matches['winner']=Matches['winner'].astype(str)
Matches['player_of_match']=Matches['player_of_match'].astype(str)
Matches['venue']=Matches['venue'].astype(str)
Matches['date']=pd.to_datetime(Matches['date'])
Matches['Year_date']=Matches['date'].dt.year
Matches['Year_date']=Matches['Year_date'].astype(int)
```

```python
Matches.drop(['umpire1','umpire2','umpire3'],axis=1)
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(y='winner',data=Matches,order=Matches['winner'].value_counts().index[0:16])
plt.xlabel("Winnig Matches",size=10)
plt.ylabel("Teams",size=10)
plt.title("Total Winning Matches By Teams",size=15)
sns.set_style("darkgrid")
for label in ax.containers:ax.bar_label(label)
plt.show()
```

```python
Matches.groupby('result').result.count()
```
