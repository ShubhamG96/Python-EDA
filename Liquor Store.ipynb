# Data Cleaning and Data Preparation

```python
Data.isnull().sum()
```

```python
sns.heatmap(Data.isnull())
```

```python
Data["store_location"]=Data["store_location"].fillna(0)
```

```python
sns.heatmap(Data.isnull())
```

```python
Data.duplicated().sum()
```

```python
#Converting Data Type
Data['Date'] = pd.to_datetime(Data['date'])
Data['Year'] = Data['Date'].dt.year
Data['Month'] = Data['Date'].dt.month
Data['county'].astype(str)
Data['category_name'].astype(str)
Data['vendor_name'].astype(str)
Data['city'].astype(str)
```

```python
#Droping Columns
data=Data.drop(columns=["invoice_and_item_number","zip_code","category","county_number","vendor_number","item_number",
"store_location","item_description","volume_sold_liters","volume_sold_gallons","date"])
```

```python
data.head(1)
```

```python
sns.catplot(x='Year', y='sale_dollars', data=data)
plt.title("Revenue by Year",size=20,color='black')
plt.xlabel("Year",size=15,color='black')
plt.ylabel("Sale in $",size=15,color='black')
sns.set_style('darkgrid')
plt.show()
```

```python
ax=sns.barplot(x='Month', y='sale_dollars',data=data)
plt.title("Revenue by Month",size=20,color='black')
plt.xlabel("Month",size=15,color='black')
plt.ylabel("Sale in $",size=15,color='black')
sns.set_style('darkgrid')
plt.show()
```

```python
#Most Liquor Category Sold
```

```python
Top_Brand_Category=Data.nlargest(10,"sale_dollars")[["sale_dollars","category_name"]].set_index("category_name").sort_values(by='sale_dollars',ascending = False )
Top_Brand_Category
plt.figure(figsize=(12,4))
Top_Brand_Category.plot(kind='bar',color='yellow',edgecolor='black')
plt.title("Most Liquor Category Sold",size=20,color='black')
plt.xlabel("category_name",size=15,color='black')
plt.ylabel("Sale in $",size=15,color='black')
sns.set_style('darkgrid')
plt.show()
```

```python
#Most Liquor Brand Sold
```

```python
Top_Brand=Data.nlargest(10,"sale_dollars")[["sale_dollars","county"]].set_index("county").sort_values(by='sale_dollars',ascending = False )
Top_Brand
plt.figure(figsize=(12,4))
Top_Brand.plot(kind='bar',color='red',edgecolor='black')
plt.title("Most Liquor Brand Sold",size=20,color='black')
plt.xlabel("county",size=15,color='black')
plt.ylabel("Sale in $",size=15,color='black')
sns.set_style('darkgrid')
plt.show()
```

```python
#Most Liquor Sold in City
```

```python
Top_City=data.nlargest(10,"sale_dollars")[["sale_dollars","city"]].set_index("sale_dollars").sort_values(by='sale_dollars',ascending = False )
Top_City
plt.figure(figsize=(10, 6))
plt.barh(Top_City['city'], Top_City.index, color='skyblue')
plt.xlabel('Sale Dollars')
plt.ylabel('City')
plt.title('Top 10 Cities by Sale Dollars')
plt.gca().invert_yaxis() 
plt.show()
```
