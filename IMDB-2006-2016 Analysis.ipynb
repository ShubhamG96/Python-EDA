```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
Data=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
Data.head()
```

```python
print("Number of Rows",Data.shape[0])
print("Number of Columns",Data.shape[1])
```

# Data Cleaning

```python
Data.isnull().sum()
```

```python
Data["Revenue (Millions)"]=Data["Revenue (Millions)"].fillna(0)
Data["Metascore"]=Data["Metascore"].fillna(0)
```

```python
sns.heatmap(Data.isnull())
```

```python
Drop=Data.dropna(axis=0)
Drop
```

```python
Dup=Data.duplicated().any()
print("Duplicates values are",Dup)
```

```python
Data.describe(include="all")
```

## Data Preparation With Visualisation

# Average Voting By Year

```python
Data[Data["Runtime (Minutes)"]>=180]["Title"]
```

```python
#Finding Average
Data.groupby("Year")["Votes"].mean().sort_values(ascending=False)
```

```python
sns.barplot(x="Year",y="Votes",data=Data)
plt.title("Votes by Year")
sns.set_style(style="darkgrid")
plt.show
```

# Average Revenue by Year

```python
Data.groupby("Year")["Revenue (Millions)"].mean().sort_values(ascending=False)
```

```python
sns.barplot(x="Year",y="Revenue (Millions)",data=Data)
plt.title("Revenue by Year")
plt.show
```

# Ratings For Each Director

```python
Data.groupby("Director")["Rating"].mean().sort_values(ascending=False)
```

```python
Top10=Data.nlargest(10,["Rating"])
Top10
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Director",y="Rating",data=Top10)
plt.title("Ratings for Director")
plt.show()
```

# Movie Titles With Runtime

```python
top10_Title=Data.nlargest(10,"Runtime (Minutes)")[["Title","Runtime (Minutes)"]]
top10_Title
```

```python
top10_Title.set_index("Title")
```

```python
sns.barplot(x="Runtime (Minutes)",y="Title",data=top10_Title)
plt.title("Movie Titles With Runtime")
plt.show()
```

# Movie Released By Year

```python
Data["Year"].value_counts(ascending=False)
```

```python
sns.countplot(x="Year",data=Data)
plt.title("Movie Released By Year")
plt.show()
```

# Top 10 Movies with Rating and Director

```python
top10_Movie=Data.nlargest(10,"Rating")[["Title","Rating","Director"]]
top10_Movie
```

```python
top10_Movie.set_index("Title")
```

```python
sns.barplot(x="Rating",y="Title",data=top10_Movie,hue="Director",dodge=False)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.title("Top 10 Movies with Rating and Director")
plt.show()
```

# Top 10 Movies By Highest Revenue

```python
Top10=Data.nlargest(10,"Revenue (Millions)")[["Title","Revenue (Millions)"]].set_index("Title")
```

```python
Top10
```

```python
sns.barplot(x="Revenue (Millions)",y=Top10.index,data=Top10)
plt.title("Top 10 Movies By Highest Revenue")
plt.show()
```
