# Data Visualization

```python
Shark_investment = data.groupby('brand_name')['deal_amount_lakhs'].sum()
top_10_companies = Shark_investment.sort_values(ascending=False).head(10)
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(top_10_companies.index, top_10_companies.values,color='black')
ax.set_xlabel('Brands')
ax.set_ylabel('Total Investment (in Lakhs)')
ax.set_title('Top 10 Brands Invested by Sharks',size=15,color='black')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```

```python
sharks = ['Anupam', 'Aman', 'Namita', 'Vineeta', 'Peyush', 'Ghazal', 'Ashneer']
total_deals = []

for shark in sharks:
    deals = len(data[data[shark.lower() + '_invested'] == 1])
    total_deals.append(deals)
   # Sorting sharks and total_deals in descending order
sorted_sharks, sorted_deals = zip(*sorted(zip(sharks, total_deals), key=lambda x: x[1], reverse=False))

fig, ax = plt.subplots(figsize=(7, 4))
ax.barh(sorted_sharks, sorted_deals,color='black')
for i, v in enumerate(sorted_deals):
    ax.text(v + 0.5, i, str(v), color='black', fontweight='bold')
ax.set_xlabel('Total Companies')
ax.set_ylabel('Shark')
ax.set_title('Companies Invested by Each Shark',size=15,color='black')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(total_deals, labels=sharks, autopct='%1.1f%%', startangle=90)
ax.set_title('Sharks Investment by Percentage',size=15,color='black')
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

```python
total_aman_deals = len(data[data['aman_invested'] == 1])
aman_max_investments = data[data["aman_invested"] == data["aman_invested"].max()][['brand_name', 'description', 'deal_amount_lakhs']].set_index('brand_name').sort_values(by='deal_amount_lakhs', ascending=False)
plt.figure(figsize=(10, 3))
ax = plt.gca()
ax.bar(aman_max_investments.index, aman_max_investments['deal_amount_lakhs'], color='orange')
ax.set_xlabel('Brand Name')
ax.set_ylabel('Deal Amount (in Lakhs)')
ax.set_title("Aman's Deals",size=15,color='black')
ax.set_facecolor('lightgray')
plt.xticks(rotation=90)
for i, value in enumerate(aman_max_investments['deal_amount_lakhs']):
    ax.text(i, value/2, str(value), ha='center', va='center', fontsize=8, rotation='vertical', color='black')
plt.show()
```
